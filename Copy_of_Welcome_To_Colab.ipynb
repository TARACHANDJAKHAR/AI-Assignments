## Data preprocessing

### Subtask:
Handle categorical variables in the `df` dataframe.


**Reasoning**:
I will first identify the categorical columns in the `df` dataframe. Then, I will use one-hot encoding to convert these columns to numerical representations and store the result in a new dataframe called `df_encoded`. Finally, I will display the first 5 rows of `df_encoded`.



```
categorical_cols = df.select_dtypes(include=['object']).columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df_encoded.head()
```

## Correlation analysis

### Subtask:
Calculate and visualize the correlation matrix for both datasets to understand the relationships between variables.


**Reasoning**:
Calculate and visualize the correlation matrix for the numerical columns of the `df` DataFrame (`num_df`).



```
correlation_matrix_df = num_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features in Housing Data')
plt.show()
```

**Reasoning**:
Calculate and visualize the correlation matrix for the `df1` DataFrame.



```
correlation_matrix_df1 = df1.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_df1, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Advertising Data')
plt.show()
```

## Distribution analysis

### Subtask:
Visualize the distribution of key variables in both datasets using histograms or kernel density plots.


**Reasoning**:
Create histograms for the numerical columns in the `num_df` DataFrame to visualize their distributions.



```
num_cols = num_df.columns
n_cols = 3
n_rows = (len(num_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(num_df[col], bins=30, edgecolor='black')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Reasoning**:
Now I'll create histograms for the `df1` DataFrame to visualize the distribution of its variables. This completes the subtask of visualizing distributions for both datasets.



```
num_cols_df1 = df1.columns
n_cols_df1 = 2
n_rows_df1 = (len(num_cols_df1) + n_cols_df1 - 1) // n_cols_df1

fig, axes = plt.subplots(n_rows_df1, n_cols_df1, figsize=(12, n_rows_df1 * 5))
axes = axes.flatten()

for i, col in enumerate(num_cols_df1):
    axes[i].hist(df1[col], bins=30, edgecolor='black', color='skyblue')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
